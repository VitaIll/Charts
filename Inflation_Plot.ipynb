```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly as py

dta = pd.read_excel("dtainf.xlsx")
dta['Unnamed: 0'] = pd.to_datetime(dta['Unnamed: 0'], format = 'yyy-%mm-%dd')
dta = dta.rename(columns={'Unnamed: 0': 'Date'})
dta.head()
```

```python
import plotly.graph_objects as go
import plotly.express as px

df = dta

fig = px.line(df, x="Date", y=df.columns,
              hover_data={"Date": "|%B %d, %Y"},
              title='Czech CPI index, 2015 Average = 100')
fig.update_xaxes(
    dtick="M3",
    tickformat="%b\n%Y"
)

fig.update_yaxes(
  title_text='CPI index')
 
fig.update_layout(autosize=False,
    height = 600,              
    width=1200,
    xaxis = dict(titlefont=dict(size=1)))

fig.show()

py.offline.plot(fig, filename= 'infplot.html')
```
